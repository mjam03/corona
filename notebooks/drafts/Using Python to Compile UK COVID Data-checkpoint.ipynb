## 1. UK Data Collection - Who Does What?

As a result, the Department for Health and Social Care (DHSC) - part of the gov - has the job of trying to collate the data from each of these sources and ensure it is consistent. My ultimate aim from collecting this data is to draw some inference from it regarding several popular 'COVID conspiracies' that have permeated throughout the last year and so I will focus on creating the following data:
 - Weekly death time series per country on which to build an 'excess deaths' metric

## 2. England & Wales (ONS)

The ONS collates statistics across a range of metrics, but for mortality data it has sole responsibility within the UK for only England & Wales.

Like all good statistics agencies, despite doing all the hard work to generate the stats, they output [their data](https://www.ons.gov.uk/peoplepopulationandcommunity/birthsdeathsandmarriages/deaths/datasets/weeklyprovisionalfiguresondeathsregisteredinenglandandwales) in an obscure and difficult to collate format through a variety of inconsistently named spreadsheets. Instead of manually iterating through each link to collte the spreadsheets, we can try to use Python - the library BeautifulSoup specifically - to make this easier for us **and** keep the data up to date as 2021 rolls on.

The code below _looks_ fairly ugly because _it is_ fairly ugly, but in the end it churns out the data we want.

```python
# hacky way described here to circumvent the ons website bouncing you because it knows you're not a human
# https://stackoverflow.com/questions/13303449/urllib2-httperror-http-error-403-forbidden
hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}

# name the urls - ONS_DEATH_STRING comically named
ONS_ROOT = 'https://www.ons.gov.uk'
ONS_DEATH_STRING = '/peoplepopulationandcommunity/birthsdeathsandmarriages/deaths/datasets/weeklyprovisionalfiguresondeathsregisteredinenglandandwales/'
# request website, parse and identify from the html only the url link elements
req = Request(ONS_ROOT+ONS_DEATH_STRING, headers=hdr)
html_page = urlopen(req)
soup = BeautifulSoup(html_page, "lxml")
links = []
for link in soup.findAll('a'):
    l = link.get('href')
    if l != None:
        links.append(l)

# now we filter through the links to only get the relevant links for what we want
ENG_WALES_XLS = []
for year in range(2010,2022):
    for l in links:
        if ONS_DEATH_STRING in l and str(year) in l and '.xls' in l:
            ENG_WALES_XLS.append(l)
```

```python
# now we know the urls to request the spreadsheets, we need to:
# - request them
# - get the weekyl deaths tab in the spreadsheet
# - parse into pandas and manipulate their to get the numbers we want
# - handle formatting changed throughout last 10 years
# - create homogenously named columns to concat each year of data together

# england/wales list of annual dfs of weekly deaths
ew_dfs = []

# iterate through each year prior to 2021
for year in range(2010,2021):
    # get file name, form URL, request spreadsheet
    file_name = [x for x in ENG_WALES_XLS if str(year) in x][0]
    file_name = ONS_ROOT + file_name
    r = requests.get(file_name)
    xl_obj = pd.ExcelFile(r.content)
    
    # find the sheet in the xls file that contains the data
    weekly_sheet_name = [x for x in xl_obj.sheet_names if 'Weekly' in x and str(year) in x][0]
    
    # exception handling for format change across years
    if year in [2011]:
        df_ew_temp = pd.read_excel(xl_obj, sheet_name=weekly_sheet_name, header=3)
        df_ew_temp = pd.concat([df_ew_temp.iloc[0], df_ew_temp.iloc[3], df_ew_temp.iloc[9]], axis=1)
    elif year in [2014]:
        df_ew_temp = pd.read_excel(xl_obj, sheet_name=weekly_sheet_name, header=2)
        df_ew_temp = pd.concat([df_ew_temp.iloc[0], df_ew_temp.iloc[3], df_ew_temp.iloc[8], df_ew_temp.iloc[9]], axis=1)
    elif year in [2016]:
        df_ew_temp = pd.read_excel(xl_obj, sheet_name=weekly_sheet_name, header=3)
        df_ew_temp = pd.concat([df_ew_temp.iloc[0], df_ew_temp.iloc[3], df_ew_temp.iloc[8]], axis=1)
    elif year in [2020]:
        df_ew_temp = pd.read_excel(xl_obj, sheet_name=weekly_sheet_name, header=4)
        df_ew_temp = pd.concat([df_ew_temp.iloc[0], df_ew_temp.iloc[3], df_ew_temp.iloc[12]], axis=1)
    else:
        df_ew_temp = pd.read_excel(xl_obj, sheet_name=weekly_sheet_name, header=3)
        df_ew_temp = pd.concat([df_ew_temp.iloc[0], df_ew_temp.iloc[3], df_ew_temp.iloc[8]], axis=1)
        
    # reformat to name cols, remove first row
    df_ew_temp = df_ew_temp.reset_index()
    df_ew_temp.columns = df_ew_temp.iloc[0]
    df_ew_temp = df_ew_temp[df_ew_temp.index != 0]
    # if later years then one col needs renamed from nan
    if year in [2016, 2017, 2018, 2019, 2020]:
        df_ew_temp = df_ew_temp.rename(columns={np.nan: df_ew_temp[np.nan][1]})
        df_ew_temp = df_ew_temp.dropna()
    df_ew_temp['Year'] = year
    
    ew_dfs.append(df_ew_temp)
```

We have now requested (almost) all the ONS data, but now need to do a bit more format cleaning - an example below opf what we currently have for each year

```python
ew_dfs[1].head()
```

We now need to handle the following:
 - Request 2021 data which is in yet another new format to report extra COVID relevant data
 - Format issue in 2014 data whereby the [ICD Coding](https://www.icd10data.com/ICD10CM/Codes) for respiratory illness changed
 - Clean up column names, format data types and concat the data

```python
# form query string for 2021 then request
file_name = [x for x in ENG_WALES_XLS if str(2021) in x][0]
file_name = ONS_ROOT + file_name
r = requests.get(file_name)
xl_obj = pd.ExcelFile(r.content)

# form df of deaths data
df_ew_21 = pd.read_excel(xl_obj, sheet_name='Weekly figures 2021', header=4)
df_ew_21 = pd.concat([df_ew_21.iloc[0], df_ew_21.iloc[3]], axis=1)
df_ew_21 = df_ew_21.reset_index()
df_ew_21.columns = df_ew_21.iloc[0]
df_ew_21 = df_ew_21[df_ew_21.index != 0]
df_ew_21 = df_ew_21.dropna()
df_ew_21 = df_ew_21[df_ew_21['Total deaths, all ages (2021)'] != ':']
df_ew_21['Year'] = 2021

# form df of respiratory deaths data
df_ew_21_r = pd.read_excel(xl_obj, sheet_name='Weekly figures by cause', header=5)
df_ew_21_r = pd.concat([df_ew_21_r.iloc[0], df_ew_21_r.iloc[5]], axis=1)
df_ew_21_r = df_ew_21_r.reset_index()
df_ew_21_r.columns = df_ew_21_r.iloc[0]
df_ew_21_r = df_ew_21_r[df_ew_21_r.index != 0]
df_ew_21_r = df_ew_21_r.dropna()
df_ew_21_r['Year'] = 2021

# merge them together to have resp and all deaths in same df
df_ew_21 = pd.merge(left=df_ew_21, right=df_ew_21_r, how='outer', on=['Year', 'Week number', 'Week ended'])
df_ew_21.columns = ['Week', 'week_end', 'Deaths', 'Year', 'RespiratoryDeaths']
```

```python
# handle 2014 by using np.where to merge 2 incomplete cols (change in ICD coding that year results in 2 diff cols)
ew_dfs[4]['All respiratory diseases (ICD-10 J00-J99)\nICD-10 v 2013 (IRIS)'] = np.where(ew_dfs[4]['All respiratory diseases (ICD-10 J00-J99)\nICD-10 v 2013 (IRIS)'] == ':', ew_dfs[4]['ICD-10 v 2010 (NCHS)'], ew_dfs[4]['All respiratory diseases (ICD-10 J00-J99)\nICD-10 v 2013 (IRIS)'])
ew_dfs[4] = ew_dfs[4].drop(columns=['ICD-10 v 2010 (NCHS)'])

# rename all cols then concat
for df in ew_dfs:
    df.columns = ['Week', 'week_end', 'Deaths', 'RespiratoryDeaths', 'Year']
df_ew_ed = pd.concat(ew_dfs+[df_ew_21])

# add country
df_ew_ed['Country'] = 'England & Wales'
# format dates
df_ew_ed['week_end'] = pd.to_datetime(df_ew_ed['week_end'])
df_ew_ed['week_start'] = df_ew_ed['week_end'] + dt.timedelta(days=-6)
# format deaths
df_ew_ed['Deaths'] = pd.to_numeric(df_ew_ed['Deaths'])
df_ew_ed['RespiratoryDeaths'] = pd.to_numeric(df_ew_ed['RespiratoryDeaths'])
```

We now have the following: a well-formatted dataframe of weekly deaths and respiratory deaths for England and Wales

### 2b. Creating a Homogenous Time Scale

We need a consistent mapping between the different country-specific data sources for {Year, Week} to {week_start, week_end} so we can appropriately combine each national time series if reported on differing dates

The DHSC stats are based on the ONS data (England & Wales) and so we will use this as default - we need to form a map

```python
# earliest records for weekly death data back to 1974 for scotland
START_DATE = dt.date(1973, 12, 29)

# define start dates and end dates
start_date_list = [START_DATE+dt.timedelta(days=7*x) for x in range(0,52*50)]
start_date_list = [x for x in start_date_list if x < (dt.date.today() + dt.timedelta(days=28))]
end_date_list = [x + dt.timedelta(days=-1) for x in start_date_list[1:]]

# create a df for them and a week count per year
df_week_map = pd.DataFrame({'week_start': start_date_list[:-1], 'week_end': end_date_list})
use_next_year = [x>=dt.timedelta(days=361) for x in [x - dt.date(x.year, 1, 1) for x in start_date_list[:-1]]]
df_week_map['use_next_year'] = use_next_year
# correction for 2014/2015
df_week_map.at[2139, 'use_next_year'] = True

df_week_map['Year'] = df_week_map['week_start'].apply(lambda x: x.year) + np.where(df_week_map['use_next_year'], 1, 0)
df_week_map['Week'] = df_week_map[['use_next_year', 'Year']].groupby(['Year']).cumcount() + 1

# drop helper col and ensure dt format is correct
df_week_map = df_week_map.drop(columns=['use_next_year'])
df_week_map['week_start'] = pd.to_datetime(df_week_map['week_start'])
df_week_map['week_end'] = pd.to_datetime(df_week_map['week_end'])

df_week_map.tail(10)
```

This will be used to bring the Scottish and Northern Irish data into line

## 3. Scotland

Time to repeat a similar process for the data provided for Scotland by the NRS. The NRS provides some more interesting data and so we will query/strip the website for the following:
 - Weekly death time series (just like above for the ONS)
 - Weekly Cause of Death (COD) time series from 2020+ (to help provide info on COVID vs others)

### 3a. Scotland Cause of Death Data

As before, first we go to the [NRS website](https://www.nrscotland.gov.uk/covid19stats) and identify the link to the most recent data file using BeautifulSoup

```python
# name the urls
NRS_ROOT = 'https://www.nrscotland.gov.uk'
NRS_DEATH_STRING = 'statistics/covid19/covid-deaths-21'
# request website, parse and identify from the html only the url link elements
req = Request(NRS_ROOT+'/covid19stats', headers=hdr)
html_page = urlopen(req)
soup = BeautifulSoup(html_page, "lxml")
links = []
for link in soup.findAll('a'):
    l = link.get('href')
    if l != None:
        links.append(l)

NRS_COD_ZIP = NRS_ROOT + [x for x in links if NRS_DEATH_STRING in x and '.zip' in x][0]
print("Scottish COD data at: {}".format(NRS_COD_ZIP))

# request the zip file and display the contents
r = requests.get(NRS_COD_ZIP)
zips = zipfile.ZipFile(io.BytesIO(r.content))
print("Files contained within the zip:")
zips.namelist()[:5]
```

Next, we identify the .csv files in this .zip file which we care about and read them into a pandas dataframe

```python
# these tables are for detailing COVID Mentions
scot_20_m_fd = [x for x in zips.namelist() if '2020' in x and 'Table 1' in x][0]
scot_21_m_fd = [x for x in zips.namelist() if '2021' in x and 'Table 1' in x][0]

# these tables breakdown underlying cod across categories
scot_20_cod_fd = [x for x in zips.namelist() if '2020' in x and 'Table 3' in x][0]
scot_21_cod_fd = [x for x in zips.namelist() if '2021' in x and 'Table 3' in x][0]
```

There are 2 bits of data we care about in relation to COD:
 - Underlying COD
 - COVID 'Mentions': Useful as this is the method used to combine stats across the UK's constituent nations, and it can enable Scotland-specific comparison of 'mentions' vs COD
 
We start with extracting the 'mentions' data across 2020-2021

```python
# grab 2020 and 2021 excel tables, extract, reformat and concat

# 2020
df_s_m_20 = pd.read_csv(zips.extract(scot_20_m_fd), header=2)[:4].T[2:][[0,3]]
df_s_m_20 = df_s_m_20.dropna()
df_s_m_20.columns = ['week_start', 'Deaths']
df_s_m_20['Week'] = df_s_m_20.index
df_s_m_20['Year'] = 2020

# 2021
df_s_m_21 = pd.read_csv(zips.extract(scot_21_m_fd), header=2)[:4].T[2:][[0,3]]
df_s_m_21 = df_s_m_21.dropna()
df_s_m_21.columns = ['week_start', 'Deaths']
df_s_m_21['Week'] = df_s_m_21.index
df_s_m_21['Year'] = 2021

# concat, add country and cause
df_s_m = pd.concat([df_s_m_20, df_s_m_21])
df_s_m['Cause'] = 'COVID Mention'
df_s_m['Country'] = 'Scotland'
df_s_m = df_s_m[['Country', 'Year', 'Week', 'week_start', 'Cause', 'Deaths']]
df_s_m['Week'] = pd.to_numeric(df_s_m['Week'])

df_s_m.tail()
```

Next, we extract the COD data for 2020-2021

```python
# 2020
df_s_cod_20 = pd.read_csv(zips.extract(scot_20_cod_fd), header=4)[9:16].T
df_s_cod_20.columns = df_s_cod_20.iloc[1].values
df_s_cod_20 = df_s_cod_20.iloc[2:]
df_s_cod_20 = df_s_cod_20[~df_s_cod_20.index.str.contains('Unnamed')]

df_s_cod_20.index.name = 'week_start'
df_s_cod_20 = df_s_cod_20.reset_index()
df_s_cod_20['Week'] = df_s_cod_20.index + 1
df_s_cod_20['Year'] = 2020

# 2021
df_s_cod_21 = pd.read_csv(zips.extract(scot_21_cod_fd), header=4)[9:16].T
df_s_cod_21.columns = df_s_cod_21.iloc[1].values
df_s_cod_21 = df_s_cod_21.iloc[2:]
df_s_cod_21 = df_s_cod_21[~df_s_cod_21.index.str.contains('Unnamed')]

df_s_cod_21.index.name = 'week_start'
df_s_cod_21 = df_s_cod_21.reset_index()
df_s_cod_21['Week'] = df_s_cod_21.index + 1
df_s_cod_21['Year'] = 2021

# concat
df_s_cod = pd.concat([df_s_cod_20, df_s_cod_21])
df_s_cod['Country'] = 'Scotland'
```

Finally, we merge them together so we have all the data in one dataframe

```python
cods = [x for x in df_s_cod if x!='Week' and x!='Year' and x!='Country' and x!='week_start']
df_s_cod = pd.melt(df_s_cod, id_vars=['Country', 'Year', 'Week', 'week_start'], value_vars=cods, var_name='Cause', value_name='Deaths')
df_s_cod = pd.concat([df_s_cod, df_s_m])

df_s_cod['Deaths'] = pd.to_numeric(df_s_cod['Deaths'].apply(lambda x: x.replace('-', '').replace(',', '').strip()))
df_s_cod['week_start'] = pd.to_datetime(df_s_cod['week_start'])

# homogenise week_start / week_end with ONS
df_s_cod = pd.merge(left=df_s_cod.drop(columns=['week_start']), right=df_week_map, how='left', on=['Year', 'Week'])
df_s_cod.tail()
```

### 3b. Scotland Weekly Death Data

Now we move on to the weekly death time series for Scotland. The NRS has a comparatively very long record of this back to 1974 until 2020. We can then use the data for COD we used above for 2021 to append on the weekly time series for 2021 so we have a series that runs right up to present day.

```python
# fortunately this file doesn't change and we have one file for the last 45+ years
r = requests.get('https://www.nrscotland.gov.uk/files//statistics/weekly-monthly-births-deaths-data/2020/dec/weekly-december-20-tab-2.zip')
zips = zipfile.ZipFile(io.BytesIO(r.content))
zips.namelist()
```

```python
# extract scottish death time series data from spreadsheet
df_s_ed = pd.read_csv(zips.extract(zips.namelist()[0]), header=3)
df_s_ed.head()
```

```python
# rename col, drop redundant rows, melt to column format
df_s_ed = df_s_ed.rename(columns={'Unnamed: 0': 'Week'}).dropna()
df_s_ed = df_s_ed[[x for x in df_s_ed.columns if 'Unnamed' not in x]]
df_s_ed = pd.melt(df_s_ed, id_vars='Week', value_vars=[x for x in df_s_ed.columns if x!='Week'], value_name='Deaths', var_name='Year')

# format data types
df_s_ed['Week'] = pd.to_numeric(df_s_ed['Week'].str[-2:])
df_s_ed['Year'] = pd.to_numeric(df_s_ed['Year'])
# remove number formatting from death numbers and '-' for zero/unknown
df_s_ed['Deaths'] = pd.to_numeric(df_s_ed['Deaths'].apply(lambda x: x.replace('-', '').replace(',', '').strip()))
# dropna for week 53s
df_s_ed = df_s_ed.dropna()
```

```python
# merge on dates for {Year, Week} pair, then chronologically sort
df_s_ed = pd.merge(left=df_s_ed, right=df_week_map, how='left', on=['Year', 'Week'])
df_s_ed = df_s_ed.sort_values(['Year', 'Week'])
df_s_ed['Country'] = 'Scotland'
df_s_ed = df_s_ed[['Country', 'Year', 'Week' ,'week_start', 'week_end', 'Deaths']]
```

Now we use the COD data to add on the 2021 weekly death data (for all deaths) to keep the series up to date so we have a historical weekly time series of death data that goes from 1974-2021

```python
df_s_ed_21 = df_s_cod[(df_s_cod.Year == 2021) & (df_s_cod.Cause == 'All')].drop(columns=['Cause'])
df_s_ed = pd.concat([df_s_ed, df_s_ed_21])
df_s_ed.tail(10)
```

## 4. Northern Ireland

And finally we deal with my beloved home country of Northern Ireland where we need to parse 2 spreadsheets stripped from the NISRA website:
 - [2010-2020 weekly death data](https://www.nisra.gov.uk/publications/historical-weekly-deaths-data)
 - [2021 weekly death data](https://www.nisra.gov.uk/publications/weekly-deaths)

### 4a. NI 2010-2020 Weekly Death Data

Thankfully, NI has a single well formatted spreadsheet we can strip and parse to get the weekly total deaths and respiratory related deaths from 2010-2020

```python
# 2010-2020 weekly data - request and form xl object
NI_URL = 'https://www.nisra.gov.uk/sites/nisra.gov.uk/files/publications/Weekly%20Deaths%20by%20Age%20and%20Respiratory%20Deaths%202011-2020.xls'
r = requests.get(NI_URL)
if r.status_code == 200:
    xl_obj = pd.ExcelFile(r.content)
else:
    print("Request failed with error code: {}".format(r.status_code))
```

```python
# iterate through tabs and strip relevant data
ni_dfs = []

for sn in xl_obj.sheet_names:
    df_ni = pd.read_excel(xl_obj, sheet_name=sn, header=2)
    
    df_ni = df_ni.dropna().iloc[:,[0,1,2,3,7]]
    df_ni.columns = ['Week', 'week_start', 'week_end', 'Deaths', 'RespiratoryDeaths']
    ni_dfs.append(df_ni)
    
# concat and format val types
df_ni = pd.concat(ni_dfs)
df_ni['week_start'] = pd.to_datetime(df_ni['week_start'], errors = 'coerce')
df_ni['week_end'] = pd.to_datetime(df_ni['week_end'], errors = 'coerce')
# there is a typo of 3030 instead of 2020 in the 2020 tab for week_start, need to correct
df_ni['week_start'] = np.where(df_ni['week_start'].isna(), df_ni['week_end'] + dt.timedelta(days=-6), df_ni['week_start'])
df_ni = df_ni.reset_index(drop=True)
# also a typo in the 2015 tab for a given week - this needs fixed
df_ni.loc[df_ni[df_ni['week_start'] == dt.datetime(2015,5,1)].index, 'week_start'] = pd.to_datetime(dt.datetime(2015,5,2))
df_ni.loc[df_ni[df_ni['week_end'] == dt.datetime(2015,5,7)].index, 'week_end'] = pd.to_datetime(dt.datetime(2015,5,8))
df_ni.loc[df_ni[df_ni['week_start'] == dt.datetime(2015,5,8)].index, 'week_start'] = pd.to_datetime(dt.datetime(2015,5,9))

df_ni['Year'] = np.where((df_ni.Week == 1) & (df_ni.week_start.dt.month==12), df_ni.week_start.dt.year + 1, df_ni.week_start.dt.year)

df_ni['Deaths'] = pd.to_numeric(df_ni['Deaths'])
df_ni['RespiratoryDeaths'] = pd.to_numeric(df_ni['RespiratoryDeaths'])
```

### 4b. NI 2021 Weekly Death Data

Now the final step to strip the newer COVID-related 2021 formatted spreadsheet

```python
# name the urls - ONS_DEATH_STRING comically named
NISRA_ROOT = 'https://www.nisra.gov.uk/publications/weekly-deaths'
NISRA_DEATH_STRING = 'Weekly_Deaths'
# request website, parse and identify from the html only the url link elements
req = Request(NISRA_ROOT, headers=hdr)
html_page = urlopen(req)
soup = BeautifulSoup(html_page, "lxml")
links = []
for link in soup.findAll('a'):
    l = link.get('href')
    if l != None:
        links.append(l)

NI_XLS = [x for x in links if NISRA_DEATH_STRING in x][0]
print("NI data at: {}".format(NI_XLS))
```

```python
# request the 2021 spreadsheet
r = requests.get(NI_XLS)
xl_obj = pd.ExcelFile(r.content)

# read, format and rename cols
df_ni_21 = pd.read_excel(xl_obj, sheet_name='Table 1', header=4)
df_ni_21 = df_ni_21.iloc[:,[0,1,2]].dropna()
df_ni_21.columns = ['Week', 'week_end', 'Deaths']

# force date types and ensure numeric data types
df_ni_21['week_end'] = pd.to_datetime(df_ni_21['week_end'], errors = 'coerce')
df_ni_21['week_start'] = df_ni_21['week_end'] + dt.timedelta(days=-6)

df_ni_21['Year'] = 2021
df_ni_21['Deaths'] = pd.to_numeric(df_ni_21['Deaths'])
df_ni_21['RespiratoryDeaths'] = np.nan
```

```python
# concat 2010-2020 and the 2021 data together now we have homogenised the cols/data formats
df_ni = pd.concat([df_ni, df_ni_21])
df_ni['Country'] = 'Northern Ireland'
df_ni['Week'] = pd.to_numeric(df_ni['Week'].apply(lambda x: str(x).replace('a','').replace('b','')))
df_ni.tail()
```

## 6. 'UK GOV' Data from the DHSC - COVID Data

Finally, we can query the official stats from that feed the UK Gov dashboard [here](https://coronavirus.data.gov.uk/). The UK Gov have started to 'modernise' and as a result have implemented a very helpful ['beta' API](https://coronavirus.data.gov.uk/details/download) for this data. As a result, we can easily retrieve the data which feeds the daily numbers burned into our retinas on a daily basis.

```python
# form query string for the metrics we care about
UKGOV_URL = 'https://api.coronavirus.data.gov.uk/v2/data?areaType=nation'

uk_gov_mets = ['newCasesByPublishDate',
               'newDeaths28DaysByDeathDate',
               'newDeaths28DaysByPublishDate',
               'newOnsDeathsByRegistrationDate',
               'newDailyNsoDeathsByDeathDate']

UKGOV_QUERY_URL = UKGOV_URL + '&metric=' + '&metric='.join(uk_gov_mets) + '&format=csv'
```

```python
# make request and if successful stick in df
r = requests.get(UKGOV_QUERY_URL)
if r.status_code == 200:
    df_dhsc = pd.read_csv(io.BytesIO(r.content))
else:
    print("Request failed with error code: {}".format(r.status_code))
```

```python
# drop and rename some cols, fix dtypes and sort
df_dhsc = df_dhsc.drop(columns=['areaType', 'areaCode']).rename(columns={'areaName': 'Country'})
df_dhsc['date'] = pd.to_datetime(df_dhsc['date'])
df_dhsc = df_dhsc.sort_values('date')
df_dhsc.tail()
```
