## 1. UK Population Estimate

Taking the UK as a whole, the population is estimated using 'stocks' and 'flows':
 - Census data at time _t_: This sets the 'stock' of people at each age group
 - Births and deaths since _t_: This gives the 'flow' of people between censuses
 - Potential immigration both ways: Another flow to be taken into account
 
Once we go within geographies then inter UK immigration then needs to be taken into account (not required here)

__Get the ONS Population Estimates__

### 1a. Population Distribution of the UK by Age

```python
df_uk_pop = pd.read_excel(xl_pop, sheet_name='MYE2 - Persons', header=4)
df_uk_pop = df_uk_pop[~df_uk_pop.Name.isna()]
df_uk_pop = df_uk_pop.rename(columns={'Geography1': 'Geography'})
df_uk_pop[df_uk_pop.Geography == 'Country']
```

```python
all_ages = [x for x in df_uk_pop.columns if x not in ['Code', 'Name', 'Geography', 'All ages']]
```

```python
df_pop_by_country = df_uk_pop[df_uk_pop.Geography == 'Country'][['Name'] + all_ages]
df_pop_by_country = df_pop_by_country[df_pop_by_country.Name.isin(['ENGLAND', 'SCOTLAND', 'WALES', 'NORTHERN IRELAND'])]
```

```python
ax = df_pop_by_country.set_index('Name').T.plot(kind='area', figsize=(18,8), cmap='Blues_r');
plt.title('UK 2019 Est Population Distribution by Constituent Country');

plt.xlabel('Age');

ax.get_yaxis().set_major_formatter(mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.ylabel('Number of People');
```

Few highlights explained well [here](https://www.ons.gov.uk/peoplepopulationandcommunity/populationandmigration/populationestimates/articles/ourpopulationwherearewehowdidwegetherewherearewegoing/2020-03-27):
 - UK exhibits population composition for a developed country with low infant mortality - i.e. not characterised by a pyramid structure
 - Population density declines >70 --> this is reflective of old people dying (the spike at the end is because those >=90 are grouped together)
 - 'Baby booms' have given rise to the bumps:
  - Bump at ~70: Post WWII baby boom, slightly diminshed vs other bumps due to life expectancy
  - Bump at ~50: 1960s baby boom due to cultural shift forward - beginning of breakdown of traditional female roles
  - Bump at ~30: 1980s/90s boom as more adults capable of children due to 1960s boom
 - Overall the populations of the constituent nations don't seem to differ much

```python
df_country_dists = df_pop_by_country.set_index('Name').T / df_pop_by_country.set_index('Name').T.sum(axis=0)

ax = df_country_dists.plot(figsize=(18,8), cmap='Purples_r');
plt.title('UK 2019 Est Population % By Individual Total Population');

plt.xlabel('Age');

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.2%}'.format(x) for x in vals])
plt.ylabel('% of People vs Each Country Population');
```

So we can confirm that the population distributions of the constituent countries are very similar
